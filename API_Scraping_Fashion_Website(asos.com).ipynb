## Put everything together - Loop through result and append data inside a list

```python
product_id = []
product_code = []
product_url = []
name = []
brand = []
current_price = []
previous_price = []
images = []


for result in result_iteams:
    
    product_id.append(result['id'])
    
    product_code.append(result['productCode'])
    
    product_url.append(base_url+result['url'])
    
    name.append(result['name'])
    
    brand.append(result['brandName'])
    
    current_price.append(result['price']['current']['text'])
    
    previous_price.append(result['price']['previous']['text'])
    
    images.append(result['imageUrl'])
```

```python
data = pd.DataFrame({'Product Id': product_id, 'Product Code': product_code, 'Product Url': product_url, 'Name': name, 'Brand': brand, 'Current Price': current_price, 'Previous Price': previous_price, 'Images Link': images})
data
```
